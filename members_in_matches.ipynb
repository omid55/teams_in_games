```python
#Omid55
import numpy as np
import collections
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
members = np.genfromtxt('members_in_matches_reduced.csv', delimiter=',', skip_header=True)
TEAM_SIZE = 5
```

```python
members_dict = collections.defaultdict(lambda: 0)
for member in members.flatten():
    if member > 0:
        members_dict[member] += 1
```

```python
allplays = np.array(list(members_dict.values()))
print('There are {0} unique players in total and maximum number of play is {1}.'.format(
        len(allplays), 
      max(allplays)
      ))

min_plays = 5
plays = allplays[np.where(allplays >= min_plays)]
print('We have {0} people who has played at least {1} times.'.format(len(plays), min_plays))
```

```python
sns.set(rc={"figure.figsize": (20, 10)})
plt.hist(plays, bins=2000)
plt.xlabel("# matches a person played")
plt.ylabel("Volume")
plt.xlim(0,50)
fig = plt.gcf()
```

```python
sns.distplot(plays, fit=norm, kde=False)
```
